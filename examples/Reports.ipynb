```python
import fairbench as fb
```

```python
test, y, scores = fb.demos.adult(predict="probabilities")
yhat = scores>0.5
test.head(3)  # test is a pandas dataframe
```

```python
print("Predicted scores:", scores)
print("Labels:", y)
```

```python
sensitive = fb.Fork(fb.categories@test[8])
sensitive
```

```python
report = fb.multireport(predictions=yhat, scores=scores, labels=y, sensitive=sensitive)
report
```

```python
fb.interactive(report)
```

```python
print(report.auc.min.explain.explain.samples)
```
